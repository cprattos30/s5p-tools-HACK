```python
import folium

# These are approximate coordinates defining a polygon around Ohio
coordinates = [
    [38.403186, -84.820312],
    [41.977523, -84.785156],
    [42.009519, -80.507812],
    [38.616870, -80.507812],
    [38.403186, -84.820312]
]

# Create a map centered on Ohio
map_ohio = folium.Map(location=[40.190737, -82.669434], zoom_start=7)

# Add a polygon with your coordinates to the map
folium.Polygon(coordinates, color="red", fill=True).add_to(map_ohio)

# Display the map
map_ohio

```
